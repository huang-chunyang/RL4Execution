```python
df = aqdatac.get_data("csinfo",row_filter="producttype='a'") 
```

```python
df = df[df['status'] == 'active']
```

```python
df
```

```python
import pandas as pd
start_date_str, end_date_str = "2019-01-01", "2019-05-01"
start_date, end_date = pd.to_datetime(start_date_str), pd.to_datetime(end_date_str)
sym_list = df['sym'].tolist()
type(sym_list)
```

```python
df_1d = aqdatac.get_data("csbar_1d", start_date, end_date, sym_list) 
```

```python
df_1d
```

```python
df_1d_trade = df_1d[df_1d['volume']!=0]
df_1d_trade
```

```python
sym_list_top100 = df_1d_trade[['sym', 'volume']].groupby('sym').mean().nlargest(100, 'volume').index
```

```python
sym_list_top100
```

```python
df_train = df_1d_trade[df_1d_trade['sym'].isin(sym_list_top100)]
df_train['date'] = pd.to_datetime(df_train['date'])
df_train
```

```python
df_train['date'][0]
```

```python
trade_days = aqdatac.get_data("csdate",start_date, end_date)
```

```python
trade_days = trade_days['date']
```

```python
trade_days = pd.to_datetime(trade_days)
```

```python
prev_day_dict = {day: trade_days[trade_days < day].max() for day in trade_days}
prev_day_dict
```

```python
df_train['prev_date'] = df_train['date'].map(prev_day_dict)
df_train_copy = df_train.copy()
```

```python
df_train_copy
```

```python
df_train
```

```python
df_train = df_train.dropna() # df1
merged_df = df_train.merge(df_train_copy[['date', 'sym', 'volume']], left_on=['prev_date', 'sym'], right_on=['date', 'sym'], how='left', suffixes=('', '_prev'))
merged_df
```

```python
result = merged_df[['date', 'sym', 'volume', 'volume_prev']].dropna()
```

```python
result.to_csv('../train_set.csv')
```
